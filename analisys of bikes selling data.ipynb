```python
import pandas as pd

dataset = pd.read_csv('Sales.csv')
bike_data = pd.DataFrame(dataset)
bike_data
```

```python
#drop unnecessary columns for better focusing on data
bike_data.drop(['Revenue', 'Age_Group', 'Unit_Cost', 'Unit_Price'], axis=1, inplace=True)
bike_data
```

```python
#Rename column
bike_data.rename(columns = {'Order_Quantity': 'Order'}, inplace=True)
```

```python
#Check data's records
bike_data.isnull()
```

```python
#Save change to csv file
bike_data.to_csv('Out_data')
```
